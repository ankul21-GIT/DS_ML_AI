## Styling plot in Seaborn

- Seaborn figure style

- Removing Axes Spines

- Scale and context

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.boxplot(y=var["total_bill"])
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
c = ["r","g","b","y"]
sns.barplot(x="day",y="total_bill",data=var,palette=c)
plt.show()
```

```python
## grid method
c = ["r","g","b","y"]
sns.barplot(x="day",y="total_bill",data=var,palette=c)
plt.grid()
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

```python
## how to change the background
sns.set_style("dark")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
## how to change the background
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

Remove axes spine

```python
## we use despine parameter to remove axis
sns.set_style("white")
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

Scale and Context

```python
# Scale

# increase and decrease figsize

sns.set_style("white")
plt.figure(figsize=(12,4))

sns.barplot(x="day",y="total_bill",data=var)

sns.despine()

plt.show()
```

```python
## content

sns.set_style("white")
sns.set_context("poster",font_scale=2)

sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python
## color


sns.barplot(x="day",y="total_bill",data=var,palette="cool")
plt.show()
```
