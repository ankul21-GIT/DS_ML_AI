```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.island, y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm",data=var)
plt.show()
```

```python
# hue parameter is used to seprate the data.
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex")
plt.show()
```

```python
## how to change the order

order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1)
plt.show()
```

```python
## hue_order 
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"])
plt.show()
```

```python
## ci parameter confidenceinterval

order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"],ci=100)
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"],ci=100,n_boot=1)
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_length_mm", y="bill_depth_mm",data=var, orient="h")
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var,color="g")
plt.show()
```

```python
## palette method is used for giving the color
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"],palette="icefire")
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"])
plt.show()
```

```python
## saturation parameter is used for reduce the color

order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"],saturation=0.4)
plt.show()
```

```python
## errcolor and how to increase errwidth

order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"], errcolor="y")
plt.show()
```

```python
## errcolor and how to increase errwidth

order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"], errcolor="y",errwidth=3)
plt.show()
```

```python
## capsize

## give size to the cap

order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"],capsize=0.6)
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"], errcolor="y",alpha=0.4)
plt.show()
```

```python
## grid and darkgrid method
sns.set(style="dark")
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"])
plt.show()
```

```python
## grid and darkgrid method
sns.set(style="darkgrid")
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=var, hue="sex", order=order1,
            hue_order=["Female","Male"])
plt.show()
```
