```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var, hue="sex")
plt.show()
```

```python
## palatte color parameter
sns.stripplot(x="day",y="total_bill",data=var, hue="sex", palette="rocket_r")
plt.show()
```

```python
## linewidth
sns.stripplot(x="day",y="total_bill",data=var, hue="sex", palette="rocket_r",linewidth=1)
plt.show()
```

```python
## edgecolor
sns.stripplot(x="day",y="total_bill",data=var, hue="sex", palette="rocket_r",linewidth=1,edgecolor="r")
plt.show()
```

```python
## jitter parameter

sns.stripplot(x="day",y="total_bill",data=var, hue="sex", palette="rocket_r",linewidth=1,
              edgecolor="r",jitter=0.4)
plt.show()
```

```python
## size parameter for increse the size
sns.stripplot(x="day",y="total_bill",data=var, hue="sex", palette="rocket_r",linewidth=1,
              edgecolor="r",size=15)
plt.show()
```

```python
# marker parameter
m={"Male":"o","female":">"}
sns.stripplot(x="day",y="total_bill",data=var, hue="sex", palette="rocket_r",linewidth=1,
              edgecolor="r",marker="<")
plt.show()
```

```python
## alpha

m={"Male":"o","female":">"}
sns.stripplot(x="day",y="total_bill",data=var, hue="sex", palette="rocket_r",linewidth=1,
              edgecolor="r",marker="<", alpha=0.4)
plt.show()
```
