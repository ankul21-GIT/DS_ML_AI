```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
## hue method

sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
## linewidth parameter

sns.violinplot(x="day",y="total_bill",data=var,linewidth=3)
plt.show()
```

```python
## palette paeameter use for giving the color


sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["lunch","dinner"],palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,palette="Dark2",saturation=0.4)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,color="r")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True)
plt.show()
```

```python
# scale parameter
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="width")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="width",inner="quart")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="width",inner="stick")
plt.show()
```

```python
## if you wanna print singl violing plot of the total bill
sns.violinplot(x=var["total_bill"])
plt.show()
```
