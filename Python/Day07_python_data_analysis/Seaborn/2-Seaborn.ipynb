```python
import seaborn as sns
import matplotlib.pyplot as plt

var1 = [1,2,3,4,5,6,7]
var2 = [2,3,4,1,5,6,7]

plt.plot(var1,var2)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var1 = [1,2,3,4,5,6,7]
var2 = [2,3,4,1,5,6,7]

x1 = pd.DataFrame({"var1":var1, "var2":var2})

sns.lineplot(x="var1", y="var2",data=x1)

plt.show()
```

```python
var1 = [1,2,3,4,5,6,7]
var2 = [2,3,4,1,5,6,7]

sns.lineplot(x=var1,y=var2)

plt.show()
```

```python
y = sns.load_dataset("penguins")
y
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y)
plt.show()
```

```python
## hue parameter
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex")
plt.show()
```

```python
## style parameter give style to any particular name
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex", style="sex")
plt.show()
```

```python
## palette method is used for giving the different color.
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex", style="sex",palette="BuGn")
plt.show()
```

```python
## marker parameter is used for giving the marker in any particular column line male and female
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex", style="sex",palette="Accent",markers=["o",">"])
plt.show()
```

```python
y1 = sns.load_dataset("penguins").head(20)
y1
```

```python
## marker parameter is used for giving the marker in any particular column line male and female
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y1,hue="sex", style="sex",palette="Accent",markers=["o",">"])
plt.show()
```

```python
## dashes romove the dashes and legen=false remove the legend method into the table
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y1,hue="sex", style="sex",palette="Accent",
             markers=["o",">"], dashes=False,legend=False)

plt.title("python")
plt.grid()
plt.show()
```
