```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
var
```

```python
sns.pairplot(var, hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill", "tip"], hue="sex")
plt.show()
```

```python
## hue_order

sns.pairplot(var,vars=["total_bill", "tip"], hue="sex", hue_order=["Female","Male"])
plt.show()
```

```python
# palette for color

sns.pairplot(var,vars=["total_bill", "tip"], hue="sex", hue_order=["Female","Male"],palette="Dark2")
plt.show()
```

```python
## x_vars parameter
sns.pairplot(var, hue="sex", hue_order=["Female","Male"],x_vars=["total_bill", "tip"])
plt.show()
```

```python
## kind parameter

sns.pairplot(var, hue="sex", hue_order=["Female","Male"],x_vars=["total_bill", "tip"],
             kind="reg") ## kde, hist, reg
plt.show()
```

```python
## kind parameter

sns.pairplot(var, hue="sex", hue_order=["Female","Male"],x_vars=["total_bill", "tip"],
             kind="kde") ## kde, hist, reg
plt.show()
```

```python
## diag_kind method
sns.pairplot(var, hue="sex", hue_order=["Female","Male"], kind="kde", diag_kind="hist") ## kde, hist, reg
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=["Female","Male"]) ## kde, hist, reg
plt.show()
```

```python
## markers method
sns.pairplot(var, hue="sex", hue_order=["Female","Male"], markers=["*","<"]) ## kde, hist, reg
plt.show()
```
