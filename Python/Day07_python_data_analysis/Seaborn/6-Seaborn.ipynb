```python
ar = np.array([['a0','a1','a2','a3','a4'],
               ['b0','b1','b2','b3','b4']])

ar
var = np.random.rand(2, 5) * 10
```

```python
sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s")
```

```python
y={"fontsize":15,"color":"r"} ## change the color and the fontsize.
sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s",annot_kws=y)
```

```python
## linewidth method increase the width between the line

y={"fontsize":15,"color":"r"} ## change the color and the fontsize.
sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s",annot_kws=y,linewidth=2)
```

```python
## linecolor metod give the color between the line
y={"fontsize":15,"color":"r"} ## change the color and the fontsize.
sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s",annot_kws=y,linewidth=2,linecolor="y")
```

```python
## cbar method remove the color bar into graph

y={"fontsize":15,"color":"r"} ## change the color and the fontsize.
sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s",annot_kws=y,linewidth=2,
            linecolor="y", cbar=False)
```

```python
## xticklabels yticklabels=False it remove the x axis and y axis parameter

y={"fontsize":15,"color":"r"} ## change the color and the fontsize.
sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s",annot_kws=y,linewidth=2,
            linecolor="y", cbar=False, xticklabels=False,yticklabels=False)

plt.show()
```

```python
## var.set method is used for the giving the label name

y={"fontsize":15,"color":"r"} ## change the color and the fontsize.
v = sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s",annot_kws=y,linewidth=2,
            linecolor="y", cbar=False, xticklabels=False,yticklabels=False)

v.set(xlabel="Python",ylabel="wscubetech")
plt.show()
```

```python
y={"fontsize":15,"color":"r"} ## change the color and the fontsize.
v = sns.heatmap(var, vmin=0, vmax=10,cmap="PuOr", annot=ar, fmt="s",annot_kws=y,linewidth=2,
            linecolor="y", cbar=False, xticklabels=False,yticklabels=False)

v.set(xlabel="Python",ylabel="wscubetech")
sns.set(font_scale=1)  ## it increase the font size 
plt.show()
```
