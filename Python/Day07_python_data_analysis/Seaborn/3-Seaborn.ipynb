```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
## createa sns histogram plot
sns.displot(var["bill_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240,250])
plt.show()
```

```python
## kde
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True)
plt.show()
```

```python
## rug parameter give density to the kde
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True, rug=True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True, rug=True,
            color="g")
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],kde=True, rug=True,color="g", log_scale=True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],kde=True,rug=True)
plt.show()
```
