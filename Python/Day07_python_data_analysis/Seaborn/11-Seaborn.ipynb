```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
## hue mathod 
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
## hue mathod 
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
## color
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g")
plt.show()
```

```python
## order parameter used for change the order
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g", order=["Fri","Sun","Thu","Sat"])
plt.show()
```

```python
## showmeans show the median value
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g", order=["Fri","Sun","Thu","Sat"],
            showmeans=True)
plt.show()
```

```python
## meanprops method 
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="w", order=["Fri","Sun","Thu","Sat"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()
```

```python
## linewidth
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,color="w", order=["Fri","Sun","Thu","Sat"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3)
plt.show()
```

```python
# palette parameter used for giving a color.
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,order=["Fri","Sun","Thu","Sat"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma")
plt.show()
```

```python
# horizontal data
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",data=var,order=["Fri","Sun","Thu","Sat"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,
            palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,
            palette="plasma",orient="h")
plt.show()
```

```python
sns.set(style="whitedrid")

sns.boxplot(x=var["total_bill"])

plt.show()
```
