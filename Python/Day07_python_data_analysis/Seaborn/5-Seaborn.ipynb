```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
var = sns.load_dataset("penguins").head(20)
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
## change the style using style parameter
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex")
plt.show()
```

```python
## size parameter

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex")
plt.show()
```

```python
## sizes parameter

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",
                sizes=[200,100])
plt.show()
```

```python
## palette parameter for colour

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",
                sizes=[200,100],palette="gist_ncar")
plt.show()
```

```python
## alpha parameter for reduce the colour

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",
                sizes=[200,100],palette="gist_ncar",alpha=0.6)
plt.show()
```

```python
## marker parameter
m={"Male":"+","Female":"o"}

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",
                sizes=[200,100],palette="gist_ncar",markers=m)
plt.show()
```
