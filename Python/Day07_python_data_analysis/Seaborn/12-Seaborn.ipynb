```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var = sns.load_dataset("tips").head(20)
var
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex")
plt.show()
```

```python
# palatte parameter
sns.catplot(x="tip",y="size",data=var,hue="sex",palette="BuPu")
plt.show()
```

```python
## height parameter used for reduce the height

sns.catplot(x="tip",y="size",data=var,hue="sex",palette="Oranges",height=3)
plt.show()
```

```python
## height parameter used for reduce the height

sns.catplot(x="tip",y="size",data=var,hue="sex",palette="Oranges",height=8)
plt.show()
```

```python
## kind parameter 
sns.catplot(x="tip",y="size",data=var,hue="sex",palette="Oranges",kind="bar")
plt.show()
```

```python
## kind parameter 
sns.catplot(x="day",y="size",data=var,hue="sex",palette="Oranges",kind="point")
plt.show()
```

```python
## kind parameter 
sns.catplot(x="day",y="size",data=var,hue="sex",palette="Oranges",kind="box")
plt.show()
```
