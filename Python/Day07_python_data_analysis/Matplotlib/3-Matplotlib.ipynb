```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["r","y","g","b","r","g","r"]

plt.scatter(day,no,color=c)
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.show()
```

```python
## how to give a different different size or change the size
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["r","y","g","b","r","g","r"]
s=[160,170,80,90,200,120,230]
plt.scatter(day,no,color=c,sizes=s)
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["r","y","g","b","r","g","r"]
s=[160,170,80,90,200,120,230]
plt.scatter(day,no,color=c,sizes=s,alpha=0.4) ## alpha method reduce the transparency
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.show()
```

```python
## marker parameter change the method in it.
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["r","y","g","b","r","g","r"]
s=[160,170,80,90,200,120,230]
plt.scatter(day,no,color=c,sizes=s,marker="*", edgecolor="g", linewidth=1) ## edgecolor method is change the color of the edge
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.show()
```

```python
## cmap method is used for giving a different different colors
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [10,49,30,29,56,20,30]
s = [160,170,80,90,200,120,230]

plt.scatter(day, no, c=colors, s=s, cmap="viridis", marker='o') 
plt.colorbar()
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.colorbar() 
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [10,49,30,29,56,20,30]
s = [160,170,80,90,200,120,230]

plt.scatter(day, no, c=colors, s=s, cmap="Accent", marker='o', alpha=0.6) 
plt.colorbar()
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.show()
```

```python
## how to give title to the color bar

import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [10,49,30,29,56,20,30]
s = [160,170,80,90,200,120,230]

plt.scatter(day, no, c=colors, s=s, cmap="Accent", marker='o') 
t = plt.colorbar()
t.set_label("Color Bar")
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [3,2,4,5,1,6,2]
colors = [10,49,30,29,56,20,30]
s = [160,170,80,90,200,120,230]

plt.scatter(day, no, c=colors, sizes=s, cmap="Accent",alpha=0.4) 
plt.scatter(day, no2, c="r", s=s, alpha=0.5) 
t = plt.colorbar()
t.set_label("Color Bar")
plt.title("Scatter", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.show()
```
