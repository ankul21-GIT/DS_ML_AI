## Text in matplotlib

- text = add text at an arbitrary location of the axes
- annotate = add an annotation, with an optional arrow at an arbitrary location of axes
- xlabel = add a label to the Axes's x-axis
- ylabel = add a label to the Axes's y-axis
- title = add title to the axes

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,6]
y = [3,1,5,2,6]

plt.plot(x,y)


plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,6]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("python", fontsize=15)
plt.xlabel("x-axis", fontsize=15)
plt.ylabel("y-axis", fontsize=15)



plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,6]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("python", fontsize=15)
plt.xlabel("x-axis", fontsize=15)
plt.ylabel("y-axis", fontsize=15)
plt.text(2,5,"python", style="italic")



plt.show()
```

```python
## bbox is used for give the facecolor
import matplotlib.pyplot as plt

x = [1,2,3,4,6]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("python", fontsize=15)
plt.xlabel("x-axis", fontsize=15)
plt.ylabel("y-axis", fontsize=15)
plt.text(2,5,"python", style="italic",bbox={"facecolor":"red"}) ## it show python in the red color box



plt.show()
```

```python
## annotate method

import matplotlib.pyplot as plt

x = [1,2,3,4,6]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("python", fontsize=15)
plt.xlabel("x-axis", fontsize=15)
plt.ylabel("y-axis", fontsize=15)
plt.text(2,5,"python", style="italic",bbox={"facecolor":"red"})

plt.annotate("python",xy=(2,2), xytext=(3,3), arrowprops=dict(facecolor="black",shrink=100))



plt.show()
```
