```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x)

plt.show()
```

```python
## Notch plot
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,notch=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,notch=True)

plt.show()
```

```python
## horizonatal using vert method
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,vert=False)

plt.show()
```

```python
## width parameter to increase the width
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,widths=0.4)

plt.show()
```

```python
## label parameter it give the label name and replace the 1 with python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,labels=["Python"],patch_artist=True) # patch_artist method fill the color.

plt.show()
```

```python
## showmeans method show the point of the mean value
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,labels=["Python"],showmeans=True) # patch_artist method fill the color.

plt.show()
```

Outlier

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,120]

plt.boxplot(x,labels=["Python"])

plt.show()
```

```python
## whis parameter join the wish to the outlier

import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,120]

plt.boxplot(x,labels=["Python"],whis=3) # patch_artist method fill the color.

plt.show()
```

```python
## sym method change the outlier shape and color

import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,120]

plt.boxplot(x,labels=["Python"], sym="g+")

plt.show()
```

```python
## boxprops method change the color of the box.

import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,120]

plt.boxplot(x,labels=["Python"], sym="g+",boxprops=dict(color="r"))

plt.show()
```

```python
## capprops method change the color of the cap.

import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,labels=["Python"], sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"))

plt.show()
```

```python
## whiskerprops method change the color.

import matplotlib.pyplot as plt

x = [10,20,30,40,50,60]

plt.boxplot(x,labels=["Python"], sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),
            whiskerprops=dict(color="g"))

plt.show()
```

```python
## flierprops change the properties of the filler

import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,120]

plt.boxplot(x,labels=["Python"], sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),
            whiskerprops=dict(color="g"),flierprops=dict(markeredgecolor="y"))

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,120]
x1 = [10,20,40,50,30,60,90]

plt.boxplot(x,labels=["Python"], sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),
            whiskerprops=dict(color="g"))

plt.boxplot(x1,labels=["Python"], sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),
            whiskerprops=dict(color="g"))

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,120]
x1 = [10,20,40,50,30,60,90]

y = [x,x1]

plt.boxplot(y,labels=["Python","c++"],
            showmeans=True, sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="g"))



plt.show()
```
