```python
import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y)

plt.show()
```

```python
## width paameter used for increase of decrease width

import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4, color = "y") ## color parameter for giving a color

plt.show()
```

```python
## width paameter used for increase of decrease width

import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
c = ["r","g","b","y"]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4,color=c)

plt.show()
```

```python
## align parameter is used for the aligning the given data like python, java, c, c++

import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
c = ["r","g","b","y"]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4,color=c,align="edge") 

plt.show()
```

```python
## align parameter is used for the aligning the given data like python, java, c, c++

import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
c = ["r","g","b","y"]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4,color=c,align="center") 

plt.show()
```

```python
## how to change edgecolor and width of the line boundary


import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
c = ["r","g","b","y"]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4,color=c,align="center", edgecolor="r", linewidth=1.4, linestyle=":") ## linestyle method give linestyle 

plt.show()
```

```python
## alpha parameter is used for the blur the color

import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
c = ["r","g","b","y"]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4,color=c,align="center", edgecolor="r", linewidth=1.4, linestyle=":",
        alpha=0.6) 

plt.show()
```

```python
## label parameter is used with plt.legend

import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
c = ["r","g","b","y"]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4,color="r",label="matplotlib")
plt.legend()
plt.show()
```

```python
## showing the two different color
import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)
plt.bar(x,y, width=0.4,color="r",label="matplotlib")
plt.bar(x,z, width=0.4,color="y",label="matplotlib1")
plt.legend()
plt.show()
```

```python
## it show two different bar chart

import matplotlib.pyplot as plt
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.4
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)

plt.bar(p,y, width=0.4,color="r",label="matplotlib")
plt.bar(p1,z, width=0.4,color="y",label="matplotlib1")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.4
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)

plt.bar(p,y, width=0.4,color="r",label="matplotlib")
plt.bar(p1,z, width=0.4,color="y",label="matplotlib1")

## how to give a name
plt.xticks(p+width,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.4
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)

plt.bar(p,y, width=0.4,color="r",label="matplotlib")
plt.bar(p1,z, width=0.4,color="y",label="matplotlib1")

## how to give a name
plt.xticks(p+width/2,x,rotation=15)  ## rotate parameter 
plt.legend()
plt.show()
```

```python
## how to create horizontal graph
import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)

plt.barh(x,y,color="r",label="matplotlib")
plt.bar(x,z,color="y",label="matplotlib1")

plt.legend()
plt.show()
```

```python
## how to create horizontal graph
import matplotlib.pyplot as plt
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width=0.4
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("wscube", fontsize=15)

plt.barh(p,y,width,color="r",label="matplotlib")
plt.barh(p1,z,width,color="y",label="matplotlib1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
