```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']

plt.pie(x,labels=y)


plt.show()
```

```python
## explode parametre

import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]


plt.pie(x,labels=y,explode=ex)


plt.show()
```

```python
## how to change the colors

import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c)


plt.show()
```

```python
## autopct method is used for finding the percentage


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True) # shadow give a shadow 

plt.show()
```

```python
## radius parameter to increase or decrease the size.


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8)

plt.show()
```

```python
## labeldistance method is used for the make the more distance from the title


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,labeldistance=1.8)

plt.show()
```

```python
## startangle metod change the degree of the angle.


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,
        labeldistance=1.8,startangle=90)

plt.show()
```

```python
## how to give the font size or title to increase it into large


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,
        labeldistance=1.8,startangle=0,textprops={"fontsize":15})

plt.show()
```

```python
## title is anticolockwise but we wanna clockwise rotation


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,
        labeldistance=1.8,startangle=0,textprops={"fontsize":15},counterclock=False)

plt.show()
```

```python
## wedgeprops : These method change into the shadow


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,
        labeldistance=1.8,startangle=0,textprops={"fontsize":15},counterclock=False,
        wedgeprops={"linewidth":4})

plt.show()
```

```python
## edgecolor these is the method which can change the color of the edge.


import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,
        labeldistance=1.8,startangle=0,textprops={"fontsize":15},counterclock=False,
        wedgeprops={"linewidth":4,"edgecolor":"r"}, center=(3,3))

plt.show()
```

```python
## plt.title method used for give the title

import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,
        labeldistance=1.8,startangle=0,textprops={"fontsize":15},counterclock=False,
        wedgeprops={"linewidth":4},rotatelabels=False)


plt.title("Ws cubetech")

plt.show()
```

```python
## plt.title method used for give the title
## plt.leged method is used to find the which element is given for find the location we give loc=1,2,3,4

import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
ex=[0.1,0.0,0.0,0.0]
c = ["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.8,
        labeldistance=1.8,startangle=0,textprops={"fontsize":15},counterclock=False,
        wedgeprops={"linewidth":4},rotatelabels=False)


plt.title("Ws cubetech")
plt.legend(loc=2)

plt.show()
```

Multiple pie chart : pie chart inside another pir chart

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
x1 = [40,30,20,10]
y = ['c','c++','java','python']

plt.pie(x,labels=y,radius=1.6)
plt.pie(x1,radius=0.6)

```

```python
## colors we can give the color

import matplotlib.pyplot as plt

x = [10,20,30,40]
x1 = [40,30,20,10]
y = ['c','c++','java','python']
c = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.6)
plt.pie(x1,radius=0.6,colors=c)

```

```python
## colors we can give the color

import matplotlib.pyplot as plt

x = [10,20,30,40]
x1 = [40,30,20,10]
y = ['c','c++','java','python']
c = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.6)
plt.pie(x1,radius=0.6,colors="w")

```

```python
## how to create a donete pie chart

import matplotlib.pyplot as plt

x = [10,20,30,40]
x1 = [40,30,20,10]

y = ['c','c++','java','python']
c = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.6)

cr = plt.Circle(xy=(0,0), radius=1, facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
