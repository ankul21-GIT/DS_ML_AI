```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,3,2,2,5]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,3,2,2,5]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = [area1,area2,area3]

plt.stackplot(x,area1,area2,area3,labels=l, colors=["r","m","y"])

plt.legend()
plt.show()
```

```python
## baseline z,sym,

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,3,2,2,5]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = [area1,area2,area3]

plt.stackplot(x,area1,area2,area3,labels=l, colors=["r","m","y"], baseline="sym")

plt.legend()
plt.show()
```

```python
## baseline z,sym,wiggle

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,3,2,2,5]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = [area1,area2,area3]

plt.stackplot(x,area1,area2,area3,labels=l, colors=["r","g","m"])

plt.title("Stackplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```
