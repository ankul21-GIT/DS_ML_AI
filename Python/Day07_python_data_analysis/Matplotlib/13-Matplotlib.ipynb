```python
## xlim is used for increase the limit
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [3,1,5,2]

plt.plot(x,y)

# plt.xticks(x,labels=["Python","C","C++","Java"])
# plt.yticks(x)

plt.xlim(0,10)

plt.show()
```

```python
## ylim is used for increase the limit
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [3,1,5,2]

plt.plot(x,y)

# plt.xticks(x,labels=["Python","C","C++","Java"])
# plt.yticks(x)

plt.xlim(0,10)
plt.ylim(0,10)

plt.show()
```
