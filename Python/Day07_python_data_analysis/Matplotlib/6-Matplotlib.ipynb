how to give a line

__ = Dashed Line

_. = Dash Dot line

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="r+")  ## r+ method give the color red and and plus sign

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="ro")  ## r+ method give the color red and and plus sign

plt.show()
```

```python
## bottome parameter line start with user input
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="r+", bottom=2) # bottom line start with 2

plt.show()
```

```python
## bottome parameter line start with user input
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="r+", bottom=5) # bottom line start with 2

plt.show()
```

```python
## basefmt method change the color
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="r+", basefmt="g")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="ro", basefmt="g", label="Python")

plt.legend(loc=2)
plt.show()
```

```python
## erientation method

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="ro", basefmt="g", label="Python",orientation="horizontal")

plt.legend(loc=2)
plt.show()
```

```python
## use_line_collection change the base color

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,4,3]

plt.stem(x,y,linefmt=":",markerfmt="ro", basefmt="g", label="Python",use_line_collection=False)

plt.legend(loc=2)
plt.show()
```
