Pandas-DataFrame And Series :

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
## Series
##A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
# crete a series from dictionary

data = {"a":1,"b":2,"c":3,"d":4,"e":5}
series = pd.Series(data)
print(series)
```

```python
data = [10,20,30,40,50]
stri = ['i','ii','iii','iv','v']
series = pd.Series(data,index=stri)
print(series)
```

```python
name = {
    "name" : "ankur",
    "age": 23,
    "sec" : 24,
    "gender": "mle",
}

series = pd.Series(name)
print(series)
```

```python
## Dataframe
## create a Dataframe from a dictionary of list

name = {
    "Name":["John","Bob","Casey","David"],
    "Age":[23,45,67,32],
    "City":["New York","California","Chicago","Houston"]

}

series = pd.DataFrame(name)
print(series)
```

```python
## Create a Data frame From a List of Dictionaries

data = [
    {"Name":"Ankur","Age":25,"country":"India","Profession":"Data Science"},
    {"Name":"Abhishek","Age":21,"country":"India","Profession":"Web Developer"},
    {"Name":"Lucy","Age":29,"country":"America","Profession":"Software Engineer"},
    {"Name":"Bob","Age":32,"country":"canada","Profession":"Machicine Learning"},
    {"Name":"Casey","Age":27,"country":"Australia","Profession":"Prompt Engineer"}
   
]

name = pd.DataFrame(data)
print(name)
```
