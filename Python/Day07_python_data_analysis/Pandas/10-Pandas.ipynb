```python
import pandas as pd
df = pd.read_csv(r"C:\\Users\\ankul\\Downloads\\NEW.csv")
df
```

```python
## replace method is uded to replace any row or column using replace and to_replcae method.
df.replace(to_replace=1, value=22)
```

```python
df.replace(to_replace='ELECTHERM', value="Python")
```

```python
df['SECURITY'] = df['SECURITY'].replace("GKW Limited", "python")
df
```

```python
## replace all the column with Java
df['SYMBOL'] = "Java"
df
```

```python
df['No.'] = 22
df
```

```python
df['SECURITY'] = df['SECURITY'].replace("Linc Limited","Java")
df
```

```python
## regex parameter to print whole element with same string or same integer.
df.replace("['A-Za-z']","Ankur",regex=True)
```

```python
import pandas as pd
df = pd.read_csv(r"C:\\Users\\ankul\\Downloads\\NEW.csv")
df
```

```python
## limit parameter define the how many time element in column should be replaced.
df.replace(1,method="ffill",limit=3)
```
