## Merge and Concat in Pandas

merge

```python
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,4],"c":[21,22,23,24]})
pd.merge(var1,var2,on="a")
```

```python
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,4],"c":[21,22,23,24]})
pd.merge(var2,var1,on="a")
```

```python
## 6 are not match to the element 4 in var1
pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var1,var2,on="a")
```

```python
## how=inner parameter it show only common element
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var1,var2,how = "inner")
```

```python
## how=outer parameter it show all element it give NaN where element is blank
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var1,var2,how = "outer")
```

```python
## left parametre show all the left side data
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var1,var2,how = "left")
```

```python
## how=right parametre show full data of the right side
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var1,var2,how = "right")
```

indicator

```python
## indicator parameter 
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var1,var2,how = "outer", indicator=True)
```

```python
## how=right parametre show full data of the right side
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})
pd.merge(var1,var2,how = "right", indicator=True)
```

```python
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"b":[21,22,23,24]})
pd.merge(var1,var2)
```

```python
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"b":[21,22,23,24]})
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
## suffixes method for change the header name
import pandas as pd
var1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
var2 = pd.DataFrame({"a":[1,2,3,6],"b":[21,22,23,24]})
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("_index","_id"))
```

```python
v1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
v2 = pd.DataFrame({"a":[1,2,3,6],"c":[21,22,23,24]})

pd.merge(v1,v2)
```

## concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,21,31,41])
sr2
```

```python
## how to concatenate sr1 and sr2
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,21,31,41])

pd.concat([sr1,sr2])
```

```python
v1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
v2 = pd.DataFrame({"c":[1,2,3,6],"b":[21,22,23,24]})

pd.concat([v1,v2], axis=1)
```

```python
v1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
v2 = pd.DataFrame({"c":[1,2,3,6],"b":[21,22,23,24]})

pd.concat([v1,v2], axis=0,join="inner")
```

```python
v1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
v2 = pd.DataFrame({"a":[1,2],"c":[21,22]})

pd.concat([v1,v2], axis=1)
```

```python
## join=inner method remove the NaN element in a 
v1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
v2 = pd.DataFrame({"a":[1,2],"c":[21,22]})

pd.concat([v1,v2], axis=1, join="inner")
```

```python
## key method merge data according to x axis and y axis.
v1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
v2 = pd.DataFrame({"c":[1,2,3,6],"b":[21,22,23,24]})

pd.concat([v1,v2], axis=1,keys=["v1","v2"])
```

```python
## key method merge data according to x axis and y axis.
v1 = pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
v2 = pd.DataFrame({"c":[1,2,3,6],"b":[21,22,23,24]})

pd.concat([v1,v2], axis=0,keys=["v1","v2"])
```

```python
## key method merge data according to x axis and y axis.
v1 = pd.DataFrame({"D":[1,2,3,4]})
v2 = pd.DataFrame({"A":[1,2,3,6],"C":[21,22,23,24]})

pd.concat([v1,v2])
```
