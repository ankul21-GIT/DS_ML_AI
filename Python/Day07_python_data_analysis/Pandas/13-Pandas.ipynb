## Join and append dataframe

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"C":[10,20,30,40],"D":[11,22,33,44]})

var1.join(var2)
```

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"C":[10,20,],"D":[11,22]})

var1.join(var2)
```

```python
## giving the index parameter
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]}, index=['a','b','c','d'])
var2 = pd.DataFrame({"C":[10,20,],"D":[11,22]})

var1.join(var2)
```

```python
## giving the index parameter
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]}, index=['a','b','c','d'])
var2 = pd.DataFrame({"C":[10,20,],"D":[11,22]},index=['a','b'])

var2.join(var1)
```

```python
## giving the index parameter
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]}, index=['a','b','c','d'])
var2 = pd.DataFrame({"C":[10,20,],"D":[11,22]},index=['a','b'])

var2.join(var1, how = 'right')
```

```python
## giving the index parameter
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]}, index=['a','b','c','d'])
var2 = pd.DataFrame({"C":[10,20,],"D":[11,22]},index=['a','b'])

var2.join(var1, how = 'outer')
```

```python
## giving the index parameter
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]}, index=['a','b','c','d'])
var2 = pd.DataFrame({"C":[10,20,],"D":[11,22]},index=['a','b'])

var2.join(var1, how = 'inner')
```

```python
## if both the column name is same we use lsuffix method to give new name to the column
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]}, index=['a','b','c','d'])
var2 = pd.DataFrame({"C":[10,20,],"B":[11,22]},index=['a','b'])

# var2.join(var1, how = 'inner') it give error because the column name is same in both var
var2.join(var1, how = 'inner', lsuffix="l1_s")
```

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]}, index=['a','b','c','d'])
var2 = pd.DataFrame({"C":[10,20,],"B":[11,22]},index=['a','b'])

var2.join(var1, how = 'inner', rsuffix="l_s")
```
